# Data PreProcessing

```python
df = pd.read_csv(target_path)
```

## Feature Extraction

```python
df = df[['id','name','overview','genres','type','created_by','poster_path','vote_average','original_language','production_companies','adult','networks']]
```

```python
df.head()
```

```python
df.dropna(inplace = True)
```

```python
# delete all the rows where  'original_language' is not in ['en','hi','es','ja','ko','de','fr']

df = df[df['original_language'].isin(['en','hi','es','ja','ko','de','fr'])]
```

```python
df.shape
```

```python
df.head()
```

```python
df['created_by'] = df['created_by'].str.replace(' ', '')
df['production_companies'] = df['production_companies'].str.replace(' ', '')
```

```python
df['created_by'] = df['created_by'].str.split(',')
df['production_companies'] = df['production_companies'].str.split(',')
df['genres'] = df['genres'].str.split(',')
```

```python
df['overview'] = df['overview'].apply(lambda x:x.split())
```

```python
df.head()
```

## EDA

```python
#create a new column by merging overview genres type created_by

df['tags'] = df['overview'] + df['genres'] + df['created_by'] + df['production_companies']
df.head()
```

```python
df['genres'] = df['genres'].apply(lambda x: ','.join(x))
df['genres'] = df['genres'].apply(lambda x: x.casefold())
df['overview'] = df['overview'].apply(lambda x: ' '.join(x))
```

```python
df['tags'] = df['tags'].apply(lambda x: " ".join(x))
df['tags'] = df['tags'].apply(lambda x: x.lower())
```

### Stemming tags

```python
from nltk.stem import PorterStemmer

porter_stemmer = PorterStemmer()
```

```python
def stem(text):
  y = []
  for i in text.split():
    y.append(porter_stemmer.stem(i))

  return " ".join(y)
```

```python
df['tags'] = df['tags'].apply(stem)
```

```python
#final df to be used further

tv_series = df.drop(columns = ['type','created_by','production_companies'])
```

```python
tv_series.head()
```

```python
tv_series.duplicated().sum()
```

```python
tv_series.drop_duplicates(keep=False,inplace = True)
```

```python
tv_series.shape
```

```python
# 'en','hi','es','ja','ko','de','fr'

replacements = {'en':'english','hi':'hindi','es':'spanish','ja':'japanese','ko':'korean','de':'german','fr':'french'}

tv_series['original_language'] = tv_series['original_language'].map(replacements).fillna(tv_series['original_language'])
```

```python
tv_series.head()
```

# Vectorization


```python
from sklearn.feature_extraction.text import TfidfVectorizer
tfid = TfidfVectorizer(stop_words= 'english')
```

```python
from sklearn.metrics.pairwise import cosine_similarity
```

```python
tfid_matrix = tfid.fit_transform(tv_series['tags'])
```

```python
similarity = cosine_similarity(tfid_matrix, tfid_matrix)
similarity
```

## Suggest Series based on Langauge and Genres sorted by rating

```python
# only suggests movies which have above certain rating based on genre and language

def ratingBased(genres,lang,rating = 6,num = 10):
  mov = tv_series[(tv_series['vote_average'] > rating) & (tv_series['genres'].str.contains(pat = genres.lower())) & (tv_series['original_language'] == lang.lower())]
  mov['poster_path'] = "https://image.tmdb.org/t/p/w500" + mov['poster_path']
  movie_list = mov['name'].head(num).tolist()
  mov_poster = mov['poster_path'].head(num).tolist()
  return ((mov['name'].head(num))+(" ")+(mov['poster_path'].head(num)))
```

```python
ratingBased('Action','eNGLish',8,7)
```

#Content Based movie recommendation

```python
def recommend(series,n=5):
  index = tv_series[(tv_series['name'] == series)].index[0]
  distances = sorted(list(enumerate(similarity[index])),reverse=True,key = lambda x: x[1])
  for i in distances[1:n+1]:
      print(tv_series['name'].iloc[i[0]])
      print(tv_series['overview'].iloc[i[0]])
      print(tv_series['genres'].iloc[i[0]])

      print("https://image.tmdb.org/t/p/w500" + tv_series['poster_path'].iloc[i[0]])
      print("")

```

```python
recommend('Stranger Things')
```
